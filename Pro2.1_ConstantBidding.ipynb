```python
import os
import numpy
import numpy as np
import pandas as pd
import random
os.getcwd()
os.chdir('D:/0124-MAAI/we_data')
path = 'D:/0124-MAAI/we_data/validation.csv'
valid = pd.read_csv(path)
budget = 6250000
record = np.zeros((301,4),dtype = np.float)
```

```python
i = 0
for const_bid in range(0,301):
    cost = 0
    impression = 0
    click = 0
    for index in valid.index:
        if cost >= budget:
            break
        else:
            bid = const_bid
            if bid >= valid['payprice'][index]:
                impression = impression+1
                cost = cost + valid['payprice'][index]
                if valid['click'][index] == 1:
                    click = click + 1
    record[i] = [const_bid,impression,click,cost]
    i = i+1
```

```python
pd_record = pd.DataFrame(record,columns=['bidprice_cons','num_impression','num_click','cost'])
pd_record['CTR'] = pd_record['num_click']/pd_record['num_impression']
pd_record = pd_record.sort_values('num_click',ascending=False)
print(pd_record)
```
